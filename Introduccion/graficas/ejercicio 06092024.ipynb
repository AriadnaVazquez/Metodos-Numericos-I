# Gráfica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
  return x - 3**(-x)
```

```python
f(-5),f(1)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-50, 50,100)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x))
plt.axhline(1, color="blue")
plt.axvline(0, color="green")
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid(color="pink")
plt.show()
```

```python
# Graficar ambas funciones.
plt.plot(x, x)
plt.plot(x,-3**(-x))
plt.axhline(1, color="blue")
plt.axvline(0, color="green")
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid(color="pink")
plt.show()
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(20,20))
plt.title('$f(x)=g(x)+h(x)$')
plt.plot(x, x)
plt.plot(x,-3**(-x))
plt.axhline(1, color="red")
plt.axvline(0, color="purple")
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid(color="grey")
plt.figtext(10,10,"En este gráfico, podemos observar las funciones g(x) el cual se encuentra marcado en azul, y h(x) marcado en naranja.")
plt.show()
```
