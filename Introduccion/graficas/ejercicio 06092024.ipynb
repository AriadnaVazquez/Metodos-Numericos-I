# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
  return x - 3**(-x)
```

```python
f(-5),f(1)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-50, 50,100)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x))
plt.axhline(1, color="blue")
plt.axvline(0, color="green")
plt.xlim(-50, 50)
plt.ylim(-50, 50)
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.grid(color="pink")
plt.show()
```
